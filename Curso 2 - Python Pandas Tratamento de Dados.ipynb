```python
bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca'] 
#Variável com os filtros
#Esses são os bairros que eu quero filtrar

selecao = dados['Bairro'].isin(bairros) 
#Variável que irá analisar os filtros de 'bairros' na coluna Bairro do df dados
#Pega a lista que eu criei (bairros) com os dados que eu quero filtrar no df e ve se os dados da coluna 'Bairro' estão la.

dados = dados[selecao] 
#variável que aplica o filtro da selecao no df dados.

dados.head()
```

```python
dados['Bairro'].drop_duplicates()
dados.head()
```

```python
grupo_bairro = dados.groupby('Bairro')
```

Teve que fazer um groupby dos bairros e armazenar em uma variável para facilitar o trabalho, pois queremos ver a média dos dados da coluna VALOR por BAIRRO. 

Se fosse usar só o df ai não daria certo. No caso: dados['Valor'].mean(). Ele faz a média de todos os valores da coluna valor sem critério algum.

Então agrupou-se todos os dados por bairro para realizarmos as análises EM RELAÇÃO aos BAIRROS, ou seja, a análise será a MÉDIA dos dados da coluna VALORES mas por BAIRRO!

Ao plotar esses números, por exemplo: média de quartos ou média do aluguel, ele segregara por BAIRRO! Muito melhor para nossa análise!!!!

```python
grupo_bairro
```

Fez um agrupamento dos bairros pelo groupby e armazenou na variável grupo_bairro

```python
for bairro, dados in grupo_bairro: 
    print(bairro)
```

O for le na coluna bairro do df dados todos os dados disponíveis e compara se eles estão dentro da variável grupo_bairro. Se estiver ele printa.

```python
for bairro, data in grupo_bairro: 
    print('{} -> {}'.format(bairro, data.Valor.mean()))
```

Nesse caso é o mesmo esquema, porém ele printa o cálculo do valor médio para cada bairro.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (14,6))
```

Q1 e Q3 representam os dados 50% longe da mediana (25% e 75% no quartil)

A ideia é deletar os outliers por se afastarem muito da mediana

```python
dados.boxplot(['Valor'])
```

```python
dados[dados['Valor'] >= 500000] #esses sao os dados fora da mediana
```

```python
valor = dados['Valor']
```

```python
Q1 = valor.quantile(.25)
Q1
```

```python
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ

print(Q1)
print(Q3)
print(IIQ)
print(limite_inferior)
print(limite_superior)
```

```python
selecao = (valor >=limite_inferior) & (valor <= limite_superior)
dados_new = dados[selecao]
dados_new
```

```python
dados_new.boxplot(['Valor'])
```

```python
dados.hist(['Valor'])
dados_new.hist(['Valor'])
```

### Identificando e removendo outliers por grupo 

```python
dados.boxplot(['Valor'], by = ['Tipo'])
```

```python
grupo_tipo = dados.groupby('Tipo')
grupo_tipo
```

```python
grupo_tipo = dados.groupby('Tipo')['Valor']
grupo_tipo
```

```python
Q1 =  grupo_tipo.quantile(.25)
Q3 =  grupo_tipo.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ


print(Q1)
print(Q3)
print(IIQ)
print(limite_inferior)
print(limite_superior)
```

```python
dados_new = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
    eh_tipo = dados['Tipo'] == tipo
    eh_dentro_limite = (dados['Valor'] >= limite_inferior[tipo]) & (dados['Valor'] <= limite_superior[tipo])
    selecao = eh_tipo & eh_dentro_limite
    dados_selecao = dados[selecao]
    dados_new = pd.concat([dados_new, dados_selecao])
```

```python
dados_new.boxplot(['Valor'], by = ['Tipo'])
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15, 7))

dados=pd.read_csv("C:/Users/lucas/Downloads/Alura/Formação Data Science/1 - Introdução/2-Python Pandas/Downloads/aluguel.csv/aluguel_amostra.csv", sep=';')
dados
```

```python
area = plt.figure()
g1 = area.add_subplot(1, 2, 1)
g2 = area.add_subplot(1, 2, 2)
```

```python
area = plt.figure()

g1 = area.add_subplot(1, 2, 1)
g2 = area.add_subplot(1, 2, 2)

grupo1 = dados.groupby('Tipo Agregado')['Valor']
label = grupo1.count().index
valores = grupo1.count().values
g1.pie(valores, labels = label, autopct='%1.1f%%')
g1.set_title('Total de Imóveis por Tipo Agregado')

grupo2 = dados.groupby('Tipo')['Valor']
label = grupo2.count().index
valores = grupo2.count().values
g2.pie(valores, labels = label, autopct='%1.1f%%', explode = (.1, .1, .1, .1, .1))
g2.set_title('Total de Imóveis por Tipo')
```
